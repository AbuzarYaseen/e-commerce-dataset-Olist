```python
orderReviews_df['review_comment_title'] = orderReviews_df['review_comment_title'].fillna('No Title')
```

```python
orderReviews_df['review_creation_date'] = pd.to_datetime(orderReviews_df['review_creation_date'])
orderReviews_df['review_answer_timestamp'] = pd.to_datetime(orderReviews_df['review_answer_timestamp'])
```

```python
orderReviews_df.head(10)
```

```python
orderReviews_df = orderReviews_df.dropna(subset=['review_comment_message'])
```

```python
orderReviews_df = orderReviews_df.reset_index(drop=True)
```

```python
orderReviews_df.duplicated().sum()
```

```python
orderReviews_df.to_csv('../data/cleaned/olist_order_reviews_dataset.csv', index=False)
```

### Initail Observations
Dataset contains 99224 order reviews rows initially.

Missing values found in review_comment_title which are fixed.

Missing values found in review_comment_message which are removed.

After cleaning dataset now has 40977 rows

No duplicated rows exist.
